```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("../data/vehicles.csv")
df.info()
df.head()
```

```python
fig = px.histogram(df, x='odometer')
fig.update_layout(width=1200, height=500)
fig.show()
```

```python
fig = px.scatter(df, x='odometer', y='price')
fig.update_layout(width=1200, height=500)
fig.show()
```

```python
fig = px.histogram(df, x='days_listed', color='condition', barmode='overlay', opacity=0.4)
fig.update_layout(width=1200, height=500)
fig.show()
```

```python
df.sample()
```

```python
fig = px.histogram(df, x='model_year', y='price', nbins=200)
fig.update_layout(width=1200, height=500)
fig.show()
```
